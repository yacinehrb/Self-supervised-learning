```python
import matplotlib.pyplot as plt
import numpy as np

demos = ['Context Enc', 'Rotation prediction', 'SimCLR']
train_top1_acc = [53.236, 79.494, 90.072]
train_top5_acc = [94.094, 99.124, 99.442]
test_top1_acc = [45.77, 79.91, 92.84]
test_top5_acc = [90.29, 99.12, 99.86]

bar_width = 0.2
index = np.arange(len(demos))

plt.figure(figsize=(10, 6))
plt.bar(index, train_top1_acc, bar_width, label='Train Top 1 Accuracy', color='blue')
plt.bar(index + bar_width, train_top5_acc, bar_width, label='Train Top 5 Accuracy', color='lightblue')
plt.bar(index + 2 * bar_width, test_top1_acc, bar_width, label='Test Top 1 Accuracy', color='green')
plt.bar(index + 3 * bar_width, test_top5_acc, bar_width, label='Test Top 5 Accuracy', color='lightgreen')

plt.xlabel('Demos')
plt.ylabel('Accuracy (%)')
plt.xticks(index + 1.5 * bar_width, demos)
plt.legend()

plt.show()
```
